```python
from PIL import Image
import os
import numpy as np

main_folder_path = "C:\\Users\\rawat\\Downloads\\Test_mango"

# Create lists to store the loaded images, labels, and seasons
image_list = []
labels = []
seasons = []

#dictionary mapping diseases to seasons
disease_season_mapping = {
    'Die Back': 'Rainy',
    'Cutting Weevil': 'Spring',
    'Powdery Mildew': 'Rainy',
    'Gall Midge': 'Summer',
    'Sooty Mould': 'Winter',
    'Anthracnose': 'Rainy',
    'Bacterial Canker': 'Fall',
    'Healthy': 'All Seasons'
}

all_seasons = ['Spring', 'Summer', 'Fall', 'Winter', 'Rainy']

for subfolder in os.listdir(main_folder_path):
    subfolder_path = os.path.join(main_folder_path, subfolder)
    
    # Check if 'subfolder' is a directory
    if os.path.isdir(subfolder_path):
        label = subfolder
        
        season = disease_season_mapping.get(label, 'Unknown')
        
        # Iterate through the files in the subfolder
        for filename in os.listdir(subfolder_path):
            if filename.endswith('.jpg'):  # Adjust file extensions as needed
                image_path = os.path.join(subfolder_path, filename)
                
                
                img = Image.open(image_path)
                
                image_list.append(img)
                
                labels.append(label)
                
                # Encode the season for this disease
                if season == 'All Seasons':
                    season_encoding = [1, 1, 1, 1, 1]  # "Healthy" has all seasons
                else:
                    season_encoding = [1 if season == s else 0 for s in all_seasons]
                
                seasons.append(season_encoding)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

image_input_shape = (224, 224, 3)  # Image dimensions
season_input_shape = (5,)  

from keras.layers import Input

image_input = Input(shape=(224, 224, 3), name='image_input')
season_input = Input(shape=(5,), name='season_input')

# Create the image branch
image_branch = layers.Conv2D(32, (3, 3), activation='relu')(image_input)
image_branch = layers.Flatten()(image_branch)
image_branch = layers.Dense(128, activation='relu')(image_branch)

# Create the season branch
season_branch = layers.Dense(64, activation='relu')(season_input)

# Combine both branches
combined = layers.concatenate([image_branch, season_branch])

output = layers.Dense(8, activation='softmax', name='output_layer')(combined)

model = models.Model(inputs=[image_input, season_input], outputs=output)

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```
