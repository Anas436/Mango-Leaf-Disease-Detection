```python
from PIL import Image
import os

main_folder_path = "C:\\Users\\rawat\\Downloads\\MangoLeafBD Dataset"

# Create lists to store the loaded images and their labels
image_list = []
labels = []

# Iterate through subfolders (disease categories)
for subfolder in os.listdir(main_folder_path):
    subfolder_path = os.path.join(main_folder_path, subfolder)
    
    # Check if 'subfolder' is a directory
    if os.path.isdir(subfolder_path):
        label = subfolder
        
        # Iterate through the files in the subfolder
        for filename in os.listdir(subfolder_path):
            if filename.endswith('.jpg'):  # Adjust file extensions as needed
                image_path = os.path.join(subfolder_path, filename)
                
                # Open and load the image
                img = Image.open(image_path)
                
                # Append the loaded image to the list
                image_list.append(img)
                # Append the label to the labels list
                labels.append(label)
```

```python
import random
import matplotlib.pyplot as plt

# Randomly select an image from the list
random_image = random.choice(image_list)

# Display the selected image
plt.imshow(random_image)
plt.axis('off')
plt.show()
```

```python
# Iterate through 'image_list' and print information about each image
for i, img in enumerate(image_list, start=0):
    # Get image dimensions
    width, height = img.size

    # Get image mode (e.g., RGB, grayscale)
    mode = img.mode

    # Print image information
    print(f"Image {i} - Dimensions: {width}x{height}, Mode: {mode}")
```

```python
# Define the target dimensions (e.g., 224x224 pixels)
target_width = 224
target_height = 224

resized_images = []

# Iterate through 'image_list' and resize each image
for img in image_list:
    # Resize the image to the target dimensions
    img = img.resize((target_width, target_height))
    
    # Append the resized image to the list
    resized_images.append(img)
```

```python
for i, img in enumerate(resized_images, start=0):
    width, height = img.size

    # Print image information
    print(f"Image {i} - Dimensions: {width}x{height}")
```

```python
import numpy as np

# Create a list to store the resized and normalized images
resized_and_normalized_images = []

for img in resized_images:

    # Convert the resized image to a NumPy array for normalization
    img_array = np.array(img)

    # Normalize pixel values to the range [0, 1]
    normalized_img = img_array / 255.0  # Assuming the original range is [0, 255]

    # Append the resized and normalized image to the list
    resized_and_normalized_images.append(normalized_img)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoded = label_encoder.fit_transform(labels)
print(label_encoded)
```

```python
from sklearn.model_selection import train_test_split

# Split your dataset into training (70%) and temporary (30%) sets
X_train_temp, X_test, y_train_temp, y_test = train_test_split(resized_and_normalized_images,label_encoded, test_size=0.3, random_state=42)

# Split the temporary set into validation (15%) and final training (15%) sets
X_train, X_val, y_train, y_val = train_test_split(X_train_temp, y_train_temp, test_size=0.5, random_state=42)
```

```python
import numpy as np

X_train_array = np.array(X_train)
X_val_array = np.array(X_val)
X_test_array = np.array(X_test)

print("Shape of X_train:", X_train_array.shape)
print("Shape of X_val:", X_val_array.shape)
print("Shape of X_test:", X_test_array.shape)
```

```python
# Confirm the shape and value range of an example image from X_train
print("Shape of an example image:", X_train[0].shape)
print("Minimum pixel value:", np.min(X_train[0]))
print("Maximum pixel value:", np.max(X_train[0]))
```

```python
# Number of classes
num_classes = len(set(labels))
print("Number of Classes:", num_classes)

# Verify the label encoding
decoded_labels = label_encoder.inverse_transform(label_encoded)
print("Sample of Decoded Labels:", decoded_labels[:10])
```

```python
print("Shape of X_train:", len(X_train))
print("Shape of y_train:", len(y_train))
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Define the model
model = Sequential()

# Add convolutional layers
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

# Flatten the output before fully connected layers
model.add(Flatten())

# Add fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(8, activation='softmax'))  # Assuming 10 classes for classification

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Check the model summary to verify architecture
model.summary()
```

```python
# Train the model on your training data
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val))

# Evaluate the model on test data to get accuracy and loss
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
print(f"Test Loss: {test_loss}")
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
